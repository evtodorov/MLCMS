```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
path = "RIMEA_scen7.txt"
data = pd.read_csv(path)
df = pd.DataFrame(data)
df['speeds'] = df.iloc[:, 0]
print(len(df), 2*'\n', df.head())
x = np.linspace(0.8, 1.85)
hist, bins = np.histogram(df['speeds'], bins=8)
plt.hist(df['speeds'], bins=bins, label = "Simulated speed")
plt.xlabel("Speed [m/s]", size = 14)
plt.ylabel("Number of pedestrians in bin", size = 14)
plt.plot(x, 7.2*norm.pdf(x, loc=1.34, scale=0.25), label = '~N(1.34, 0.25)')
plt.legend()
```
