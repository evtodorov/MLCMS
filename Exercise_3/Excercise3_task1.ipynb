```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def phase_plot(X, Y, U, V, pname):
    '''
    Creates the streamplot and quiver plot for the given
    dynamical system
    '''
    plt.streamplot(X, Y, U, V, density = 0.5)
    plt.quiver(X, Y, U, V)
```

```python
def diff_calculator(vec_x,alpha):
    '''
    Calculates the differential operator by using th formula given
    A_alpha = [[alpha, alpha],
               [-0.25,   0  ]]
    '''
    A_alpha = np.array([[alpha, alpha],[-0.25, 0]])
    #A_alpha = np.array([[1, 0],[0, 1]])
    
    return np.matmul(A_alpha,vec_x)
```

```python
X,Y = np.meshgrid( np.arange(-1.0,1.0,0.1),np.arange(-1.0,1.0,0.1) )
vec_x = np.array([X.flatten(),Y.flatten()])

alpha = 0.1

vec_velocity = diff_calculator(vec_x,alpha)

U = vec_velocity[0].reshape(X.shape)
V = vec_velocity[1].reshape(Y.shape)


phase_plot(X,Y,U,V,"Ex3_Task1_a")
```
