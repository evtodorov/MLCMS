```python
alpha_plot = np.linspace(-1, 1, 100)
branch1 = np.sqrt(alpha_plot)
branch2 = -np.sqrt(alpha_plot)
plt.figure()
plt.scatter(alpha_plot, branch1, label='stable')
plt.scatter(alpha_plot, branch2, label='unstable')
plt.legend()
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$x_\infty$")
plt.xlim([-1, 1])
plt.grid()
```

```python
def task2_1t(alpha=1.):
    T, Y = np.meshgrid( np.arange(0,2,0.1), np.arange(-2,2,0.1))
    U = np.ones(T.shape)
    V = alpha - Y**2
    plt.figure()
    plt.streamplot(T, Y, U, V)
    plt.xlabel(r'$t$')
    plt.ylabel(r'$x$')
interact(task2_1t)
```

```python
def task2_1x(alpha=1.):
    X, Y = np.meshgrid( np.arange(-2,2,0.1), np.arange(-2,2,0.1))
    U = alpha - X**2
    V = -Y
    plt.figure()
    plt.streamplot(X, Y, U, V)
    plt.xlabel(r'$x$')
    plt.ylabel(r'$x_2 = - \dot x_2$')
interact(task2_1x)
```

```python
alpha_plot2 = np.linspace(0, 5)
lam7_1 = sy.lambdify(alpha, sol7.args[0], 'numpy')
lam7_2 = sy.lambdify(alpha, sol7.args[1], 'numpy')
plt.figure()
plt.scatter(alpha_plot2, lam7_1(alpha_plot2), label='stable')
plt.scatter(alpha_plot2, lam7_2(alpha_plot2), label='unstable')
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$x_\infty$")
plt.xlim([0, 5])
plt.grid()
plt.legend()
```
