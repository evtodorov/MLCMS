```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D
```

```python
def lorenz(t, X, sigma, beta, rho):
    #The Lorenz equations
    u, v, w = X
    up = -sigma*(u - v)
    vp = rho*u - v - u*w
    wp = -beta*w + u*v
    return up, vp, wp
```

## Lorenz System with initial value (10,10,10)

```python
# Figure sepcifications
WIDTH = 1000
HEIGHT = 750
DPI = 100

#Lorenz attracor values
t_end = 1000
N = 100000
sigma_val = 10
beta_val = 8/3
rho_val = 28

#initializing values
u0, v0, w0 = 10,10,10
x0 = (u0,v0,w0)

# Integrate the Lorenz equations.
soln = solve_ivp(lorenz, (0, t_end), x0, args=(sigma_val, beta_val, rho_val),dense_output=True)

# Interpolate solution onto the time grid, t.
t = np.linspace(0, t_end, N)
x, y, z = soln.sol(t)
```

```python
# Plot the Lorenz attractor using a Matplotlib 3D projection.
fig = plt.figure(figsize=(WIDTH/DPI, HEIGHT/DPI))
ax = fig.gca(projection='3d')
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

s = 10
cmap = plt.cm.winter
for i in range(0,N-s,s):
    ax.plot(x[i:i+s+1], y[i:i+s+1], z[i:i+s+1], color=cmap(i/N), alpha=0.4)
    
#ax.scatter(x,y,z,c=t,cmap='coolwarm')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.savefig('lorenz_basic_1.png', dpi=DPI)
plt.show()
```

## Lorenz system (basic) plotted till change > 1

```python
# Figure sepcifications
WIDTH = 1000
HEIGHT = 750
DPI = 100

#Lorenz attracor values
t_end = 1000
N = 100000
sigma_val = 10
beta_val = 8/3
rho_val = 28

#initializing values
u0, v0, w0 = 10,10,10
x0 = (u0,v0,w0)

u1,v1,w1 = (10+10**-8),10,10
x1 = (u1,v1,w1)

# Integrate the Lorenz equations.
soln_0 = solve_ivp(lorenz, (0, t_end), x0, args=(sigma_val, beta_val, rho_val),dense_output=True)
soln_1 = solve_ivp(lorenz, (0, t_end), x1, args=(sigma_val, beta_val, rho_val),dense_output=True)

# Interpolate solution onto the time grid, t.
t = np.linspace(0, t_end, N)
x_0, y_0, z_0 = soln_0.sol(t)
x_1, y_1, z_1 = soln_1.sol(t)

for i in range(N):
    dist_var = np.sqrt(((x_0[i]-x_1[i])*(x_0[i]-x_1[i]))+((y_0[i]-y_1[i])*(y_0[i]-y_1[i]))+((z_0[i]-z_1[i])*(z_0[i]-z_1[i])));
    if(dist_var > 1):
        N_limit = i;
        break;
N_limit
```

```python
# Plot the Lorenz attractor using a Matplotlib 3D projection.

#%matplotlib interactive
fig = plt.figure(figsize=(WIDTH/DPI, HEIGHT/DPI))
ax = fig.gca(projection='3d')
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

s = 10
cmap = plt.cm.get_cmap('Reds', 12)
for i in range(0,N_limit-s,s):
    ax.plot(x_0[i:i+s+1], y_0[i:i+s+1], z_0[i:i+s+1], color=cmap(i/N_limit), alpha=0.4)

#fig = plt.hold(True)
    
cmap = plt.cm.get_cmap('Blues', 15)
for i in range(0,N_limit-s,s):
    ax.plot(x_1[i:i+s+1], y_1[i:i+s+1], z_1[i:i+s+1], color=cmap(i/N_limit), alpha=0.4)
    
#ax.scatter(x,y,z,c=t,cmap='coolwarm')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor difference")

plt.savefig('lorenz_basic_diff.png', dpi=DPI)
plt.show()
```
