```python
def logistic_map(r, x):
    return r*x*(1-x)
```

```python
xs = []
rs = np.linspace(0,4,4000)
rplots = []
for r in rs:
    x0s = np.random.random(10)
    for x0 in x0s:
        x = x0
        for i in range(101):
            xs.append(x)
            rplots.append(r)
            x = logistic_map(r, x)
plt.figure()
plt.scatter(rplots, xs, s=0.001, alpha=0.1)
plt.xlabel(r'$r$')
plt.ylabel(r'$x$')
```
