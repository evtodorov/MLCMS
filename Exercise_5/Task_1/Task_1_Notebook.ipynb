```python
import numpy as np
import matplotlib.pyplot as plt
from helper_functions import *
```

## First part
#### Load data and visualise it

```python
x_es1, f1 = load_to_numpy('linear_function_data.txt')
plt.scatter(x_es1, f1)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('x', fontsize = 30) 
plt.ylabel(r'$f(x)$', fontsize = 30)
plt.show()
```

#### Fit the loaded data

```python
f_hat1 = linear_fit(x_es1, f1)
```

#### Visualise the fit and the original data together

```python
plt.scatter(x_es1, f1, label = r"Original data, $f(x)$", s=8)
plt.plot(x_es1, f_hat1, label = r"Fitted function, $\hat{f}(x)$", c = 'orange')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('x', fontsize = 30) 
plt.ylabel(r'$f(x)$ and $\hat{f}(x)$', fontsize = 30)
plt.legend(prop={'size': 12})
plt.show()
```
