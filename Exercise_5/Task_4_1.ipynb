```python
time = [i for i in range(data.shape[0])]
t_plot = np.array(time)
p = -1
for ii in range(data.shape[0]-1):
    if abs(data[ii+1,0]-data[0,0]) < 0.001:
        p = ii;
        break
print("The period = ")
print(p)
plt.plot(t_plot,data[:,0])
```
