```python
import gensim
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.preprocessing
from sklearn.manifold import SpectralEmbedding
```

```python
#from pmlb import fetch_data
#adult = fetch_data('adult', local_cache_dir='pmlb/')
```

```python
data = np.loadtxt('../Exercise_4/Task_2/data_DMAP_PCA_vadere.txt')
print(f'Data loaded with shape {np.shape(data)}')
```

```python
embedding = SpectralEmbedding(n_components=3)
spectral_data = embedding.fit_transform(data)
spectral_data.shape
```

```python
embedding.get_params()
```

```python
plt.scatter(spectral_data.T[0], spectral_data.T[1], s=5)
plt.show()
```

```python
fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(211, projection='3d')

ax.scatter(spectral_data.T[0], spectral_data.T[1], spectral_data.T[2], c='orange')
ax.view_init(50, 160)
plt.show()
```
